### Exercise 2

Produce a dispersion plot of the four main protagonists in Sense and Sensibility: Elinor, Marianne, Edward, and Willoughby. What can you observe about the different roles played by the males and females in this novel? Can you identify the couples?

### Exercise 5

What is the difference between the following two lines? Calculate the two values:

    len(sorted(set(w.lower() for w in text1)))
    len(sorted(w.lower() for w in set(text1)))

In the first line: lower() method converts all the words in the string into lowercase. set() method is used to get rid of the duplicates and sorted() function sorts them and lastly the len() function counts the number of unique lowercase words in the text1.

In the second line the ordering of the functions is different. We first get rid of the duplicates and then convert words into lowercase, and then finally count them. 

```python
from nltk.book import text1
value1 = len(sorted(set(w.lower() for w in text1)))
value2 = len(sorted(w.lower() for w in set(text1)))

print("Value 1:", value1)
print("Value 2:", value2)
```

### Exercise 7

Find all the four-letter words in `text6`. With the help of a frequency distribution (`FreqDist`), show these words in decreasing order of frequency.


```python
from nltk.book import text6
from nltk import FreqDist

# all the four-letter words in text6
four_letter_words = [word.lower() for word in text6 if len(word) == 4 and word.isalpha()]

# frequency distribution for the four-letter words
freq_dist = FreqDist(four_letter_words)

# the four-letter words in decreasing order of frequency
sorted_freq_dist = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
for word, frequency in sorted_freq_dist:
    print(f"{word}: {frequency}")
```

### Exercise 8

Create a set for the words in `text6`. Use a `for` and an `if` statement to loop over the words in the set and print all titlecased words with more than one character, one per line.


```python
from nltk.book import text6

word_set = set(text6)

for word in word_set:
    if word.istitle() and len(word) > 1:
        print(word)
```

### Exercise 10

### Exercise 11

What does the following Python code do? 

`sum(len(w) for w in text1)`

Can you use it to work out the average word length of `text1`?

 -> That line of Python code calculates the sum of the lengths of all words in text1. If we divide this calculation by the number of total words in text1, we'll get average word length.

```python
average_word_length = sum(len(w) for w in text1) / len(text1)
print(average_word_length)
```

### Exercise 12

Define a function `freq(word, text)` that calculates how often a given word occurs in a text, not using `count()` but a `FreqDist`. Use the function to calculate how often "promise" appears in `text4`.

```python
from nltk.book import text4

def freq(word, text):
    fdist = nltk.FreqDist(text)
    return fdist[word]

word_to_count = "promise"
frequency_in_text4 = freq(word_to_count, text4)

print("Frequency of 'promise' in text4:", frequency_in_text4)
```
